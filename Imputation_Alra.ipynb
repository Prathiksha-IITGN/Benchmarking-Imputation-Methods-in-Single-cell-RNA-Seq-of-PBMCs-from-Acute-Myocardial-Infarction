```python
# Notebook 03: Imputation with scVI

import scanpy as sc
import scvi
import os
```

```python
# -----------------------------
# Paths
# -----------------------------
input_file = "dropout_h5ad/adata_dropout_mf10_run1.h5ad"
output_file = "Imputed_h5ad/adata_mf10_scvi_imputed.h5ad"

# Ensure output directory exists
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# -----------------------------
# Load Dropout / Missing Data
# -----------------------------
adata = sc.read_h5ad(input_file)

# -----------------------------
# Setup scVI
# -----------------------------
scvi.model.SCVI.setup_anndata(adata)
```

```python

# -----------------------------
# Train the model
# -----------------------------
model = scvi.model.SCVI(adata)
model.train(max_epochs=400)  # you can adjust epochs if needed

# -----------------------------
# Get normalized (imputed) expression
# -----------------------------
adata.X = model.get_normalized_expression()

# -----------------------------
# Save imputed data
# -----------------------------
adata.write(output_file)
print(f"Saved scVI imputed data as {output_file}")
```
